반복문과 조건문 함께사용

```python
scores=[42,60,77,89,20]

for i in scores:
    if i>=60:
        print(f'{i}점은 합격입니다')
    else:
        print(f'{i}점은 불합격입니다')
```

```python
numbers=[1,2,3,4,5]
total=0
#합계: numebers를 다 더한값

for i in numbers:
    total+=i
print('합계:',total)
```

```python
scores=[65,82,91,100,58,77]
total=0
#70점 이상은 합격
#합격자는 x명 입니다. <--출력

for i in scores:
    if i>=70:
        total+=1
print(f'합격자는 {total}명 입니다')
```
