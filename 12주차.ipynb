# 여러 변수를 사용한 회귀 모델

실제 시험 점수는 공부시간만으로 결정되지 않음
- 공부시간
- 출석률
- 과제점수

> 입력 변수가 하나면 **단순 회귀** 입력 변수가 여러 개이면 **다중 회귀**

```python
import pandas as pd
import matplotlib.pyplot as plt

#머신러닝 도구
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
student_df=pd.DataFrame({
    "study_hours":[1,2,3,4,5],
    "attendance":[10,20,30,40,50],
    "assignment":[10,20,30,40,50],
    "scores":[60,70,80,90,100],
})
```

## 단순 모델 생성

```python
X=student_df[['study_hours']]
y=student_df['scores']

print("X모양: ",X.shape)
print("y모양: ",y.shape)
```

```python
X_multi=student_df[["study_hours","attendance","assignment"]]
y_multi=student_df["scores"]

X_train,X_test,y_train,y_test=train_test_split(
    X,y,
    test_size=0.3,
    random_state=42
)

multi_model=LinearRegression()
multi_model.fit(X_train,y_train)

model=LinearRegression()
model.fit(X_train,y_train)

print("모델학습완료")

pred=model.predict(X_test)

print("예측 완료")
```

## 다중 회귀 모델의 개수 확인하기

선형회귀는 입력 변수이다 하나의 개수를 학습한다.<br>
개수는 다른 조건이 같을때 해당 변수가 +1 되면 예측값이 얼마나 변하는지를 의미
단, 실제 해석은 데이터 단위와 변수 간 관계을 함께 고려해야 합니다.

```python
coef_df=pd.DataFrame({
    'feature':X_multi.columns,
    'coefficient':multi_model.coef_
})
print("절편",multi_model.intercept_)
coef_df
```

## 단순 회귀 다중 회귀 성능 비교하기

```python
multi_pred=multi_model.predict(X_test)

simple_mae=mean_absolute_error(y_test,pred)
simple_r2=r2_score(y_test,pred)

multi_mae=mean_absolute_error(y_test,multi_pred)
multi_r2=r2_score(y_test,multi_pred)

compare_df=pd.DataFrame({
    "model":['study_hours_only','study_hours + attendance + assginment'],
    'MAE':[simple_mae,multi_mae],
    'R2':[simple_r2,multi_r2],
})

compare_df
```
